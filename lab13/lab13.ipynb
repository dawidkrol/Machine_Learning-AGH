```python
import pandas as pd

df = pd.read_csv('datasets/hour.csv',
                parse_dates={'datetime': ['dteday', 'hr']},
                date_format='%Y-%m-%d %H',
                index_col='datetime'
                )
```

```python
from sklearn.metrics import mean_absolute_error

diff_7 = df[['cnt']].diff(7)['cnt']
mae_weekly = diff_7.abs().mean()*10**3/7
mae_weekly
```

```python
diff_1 = df[['cnt']].diff(1)['cnt']
mae_daily = diff_1.abs().mean()*10**3
mae_daily
```

```python
cnt_train = df['cnt']['2011-01-01 00:00':'2012-06-30 23:00']
cnt_valid = df['cnt']['2012-07-01 00:00':]
```

```python
seq_len = 1 * 24
train_ds = tf.keras.utils.timeseries_dataset_from_array(
    cnt_train.to_numpy(),
    targets=cnt_train[seq_len:],
    sequence_length=seq_len,
    batch_size=32,
    shuffle=True,
    seed=42
    )
valid_ds = tf.keras.utils.timeseries_dataset_from_array(
    cnt_valid.to_numpy(),
    targets=cnt_valid[seq_len:],
    sequence_length=seq_len,
    batch_size=32
    )
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[seq_len])
    ])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.experimental.SGD(momentum=0.9), 
    loss=tf.keras.losses.MeanSquaredError(),
    metrics=["mae"])
```

```python
history = model.fit(train_ds, epochs=20, validation_data=valid_ds)
```

```python
val_mae = tuple([model.evaluate(valid_ds, verbose=0)[1] * 10**3,])

with open('mae_linear.pkl', 'wb') as handle:
    pickle.dump(val_mae, handle)

model.save('model_linear.h5')
```

```python
data_train = df[['cnt','weathersit','atemp','workingday']]['2011-01-01 00:00':'2012-06-30 23:00']
data_valid = df[['cnt','weathersit','atemp','workingday']]['2012-07-01 00:00':]
```

```python
seq_len = 1 * 24
train_data = tf.keras.utils.timeseries_dataset_from_array(
    data_train.to_numpy(),
    targets=data_train['cnt'][seq_len:],
    sequence_length=seq_len,
    batch_size=32,
    shuffle=True,
    seed=42
    )
valid_data = tf.keras.utils.timeseries_dataset_from_array(
    data_valid.to_numpy(),
    targets=data_valid['cnt'][seq_len:],
    sequence_length=seq_len,
    batch_size=32
    )
```

```python
model_wide = tf.keras.Sequential([
    tf.keras.layers.SimpleRNN(32, input_shape=[None, 4]),
    tf.keras.layers.Dense(1)
    ])

model_wide.compile(
    optimizer=tf.keras.optimizers.experimental.SGD(momentum=0.9),
    loss=tf.keras.losses.MeanSquaredError(),
    metrics=["mae"])

history_wide = model_wide.fit(train_data, epochs=20, validation_data=valid_data)
```

```python
val_mae = tuple([model_wide.evaluate(valid_data, verbose=0)[1] * 10**3,])

with open('mae_rnn_mv.pkl', 'wb') as handle:
    pickle.dump(val_mae, handle)

model_wide.save('model_rnn_mv.h5')
val_mae
```
