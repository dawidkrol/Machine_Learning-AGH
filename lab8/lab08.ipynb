```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
```

```python
def plot_show(X):
    plt.figure()
    ax = plt.axes(projection='3d')
    ax.scatter3D(X[:,0], X[:,1], X[:,2], c=X[:,2])
    plt.show()
```

```python
from sklearn import datasets
data_breast_cancer = datasets.load_breast_cancer()
pca_cancer = PCA(n_components=0.9)
data_breast_cancer_PCA = pca_cancer.fit_transform(data_breast_cancer.data)
plot_show(data_breast_cancer.data)
```

```python
scaler_beast = StandardScaler()
data_breast_cancer_transformed = scaler_beast.fit_transform(data_breast_cancer.data)

pca_cancer_transformed = PCA(n_components=0.9)
data_breast_cancer_transformed_PCA = pca_cancer_transformed.fit_transform(data_breast_cancer_transformed)

plot_show(scaler_beast.fit_transform(data_breast_cancer_transformed))
```

```python
from sklearn.datasets import load_iris
data_iris = load_iris()

pca_iris = PCA(n_components=0.9)
data_iris_PCA = pca_iris.fit_transform(data_iris.data)

plot_show(data_iris.data)
```

```python
scaler_iris = StandardScaler()
data_iris_transformed = scaler_iris.fit_transform(data_iris.data)

pca_iris_transformed = PCA(n_components=0.9)
data_iris_transformed_PCA = pca_iris_transformed.fit_transform(data_iris_transformed)

plot_show(scaler_iris.fit_transform(data_iris_transformed))
```

```python
import pickle

def saveToFile(ll, filename):
    with open(filename, 'wb') as handle:
        pickle.dump(ll, handle)
```

```python
# data_breast_cancer
saveToFile(pca_cancer_transformed.explained_variance_ratio_, 'pca_bc.pkl')
```

```python
# data_iris
saveToFile(pca_iris_transformed.explained_variance_ratio_, 'pca_ir.pkl')
```

```python
ll_cancer = list()
for i in pca_cancer_transformed.components_:
    ll_cancer.append(np.argmax(i))
saveToFile(ll_cancer, 'idx_bc.pkl')
```

```python
ll_iris = list()
for i in pca_iris_transformed.components_:
    ll_iris.append(np.argmax(i))
saveToFile(ll_iris, 'idx_ir.pkl')
```
